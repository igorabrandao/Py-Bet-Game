**IMD Bet Game**

In the IMD bet game, your next move depends on the number of eyes you throw with the dice.

**Rules**

*Input:*
- Floor "f", where step = max(0,f);
- Epoch (roll) = 100;
- Seed;
- Number of rounds = **500**.

*For each rounds*

*For each epoch (roll):*

- Roll the dice;
- Determine the next step;

| Case | Whats happen |
| --- | --- |
| `Case 1,2:` | Decrease step by one |
| `Case 3,4,5:` | Increase step by one |
| `Case 6:` | Roll de dice again and move accordling to result of the new dice |

- There is a 0,1% probability that assign the step to 0 (zero) in each step;
- Store the current step

Round[i] = steps

## Data structure

Here is the section that I defined the data structure

```python
# Import the math package
import math

# Import the random package
import random

# Import the numpy package
import numpy as np

# Seed for random number generation
seed = 1
np.random.seed(seed)

##################### Program settings #####################

# You can change these parameters to simulate new results
min_floor = 0
max_floor = 100
epoch = 10
n_rounds = 50
step = 0
diceFace = 1

############################################################

# Auxiliary variables
current_floor = np.random.randint(min_floor, max_floor)
aux_floor = current_floor
current_round = 0
current_epoch = 0
aux_step = step
assign_zero = 0

# Create the program data structure
build = np.array(np.random.randint(1, 20))
rounds = [None] * n_rounds
```

## Print settings function

This function print the used parameters to run the current simulation

```python
# Function to print settings
def printSettings():
    print('##################### Program settings #####################')
    print('min_floor \t= ', min_floor)
    print('max_floor \t= ', max_floor)
    print('epoch \t\t= ', epoch)
    print('n_rounds \t= ', n_rounds)
    print('step \t\t= ', step)
    print('diceFace \t= ', diceFace)
    print('\nStarting floor \t= ', current_floor)
    print('############################################################\n')
```

## Roll dices function

This function determines what does mean rolling dices

```python
# Roll dices function
def rollDice():
    return np.random.randint(1, 7)
```

## Cases mapping

As python doesn't have *switch/case* structure, so I'm going to map the behavior
accordling to the dice face

```python
# Define the function blocks
def case12():
    global step
    step -= 1

def case345():
    global step
    step += 1

def case6():
    global step
    diceFace = rollDice()
    step += diceFace
    # print ('New Dice face: ', diceFace)

# Map the dice face to the function blocks
nextStep = {1 : case12,
            2 : case12,
            3 : case345,
            4 : case345,
            5 : case345,
            6 : case6,}
```

## Iterations

Ok, now that we have the necessary info's, let's roll some dices!

```python
# Print the current settings
printSettings()

# For each rounds
for current_round in range(current_round, n_rounds): 
    print ('\n### ROUND[', current_round, '] ###\n')
    
    # For each epoch (roll)
    for current_epoch in range(current_epoch, epoch):
        # Roll the dice
        diceFace = rollDice()
        
        # Determine the next step
        nextStep[diceFace]()
        aux_step += step

        # Change the current floor
        aux_floor = current_floor + step 
        if ( aux_floor >= min_floor and aux_floor <= max_floor ):
            current_floor += step

        # Probability to step to 0 floor
        assign_zero = (0, current_floor)[np.random.randint(1) < 0.1]

        # Print a rolling log
        print ('Dice[', current_epoch,'] face: ', diceFace, '\tSTEP: ', step, '\tFLOOR: ', current_floor, 
               '\tASSIGN_ZERO: ', assign_zero)
        
        # Increase the epoc iterator
        current_epoch += 1
        
        # Renew step result
        step = 0
    
    # Renew the epoc iterator
    current_epoch = 0
    
     # Add the step in the current round
    rounds[current_round] = aux_step
    
    # Renew tha aux_step variable
    aux_step = 0
    
    # Increase the round iterator
    current_round += 1
```
